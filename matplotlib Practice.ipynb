```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1)

plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')

plt.figure(2)

plt.subplot(211)
plt.plot(-1*t1, f(t1), 'bo', -1*t2, f(t2), 'k')

plt.subplot(212)
plt.plot(-1*t2, np.cos(2*np.pi*t2), 'r--')

plt.show()
```

### Text

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

n, bins, patches = plt.hist(x, 50, normed = 1, facecolor = 'r', alpha = 0.75)

plt.xlabel("Smarts", fontsize = 14, color = 'red')
plt.ylabel("Probablility")
plt.title("Histogram of IQ")
plt.text(45, 0.026, r'$\mu_{data} = 100,\ \sigma_{data}=15$', fontsize = 12, color = 'red')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

### Annotating Text

```python
ax = plt.subplot(111)
t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = plt.plot(t, s, lw = 2)

plt.annotate('local max', xy = (2, 1), xytext = (3, 1.5), 
             arrowprops = dict(facecolor = 'black', shrink = 0.05))

plt.ylim(-2, 2)
plt.show()
```

### Logarithmic Scales

```python
x = np.arange(20)
plt.figure(1)

plt.subplot(221)
plt.plot(x, x, 'ro')
plt.xscale('log')
plt.yscale('log')
plt.grid(True)

plt.subplot(222)
plt.plot(x, x, 'ro')
plt.yscale('log')
plt.grid(True)

plt.subplot(223)
plt.plot(x, x, 'ro')
plt.xscale('log')
plt.grid(True)

plt.subplot(224)
plt.plot(x, x, 'ro')
plt.grid(True)

plt.show()
```
